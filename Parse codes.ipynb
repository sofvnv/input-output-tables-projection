```python
'''About Other(NONCOMPARABLE IMPORTS AND REST-OF-THE-WORLD ADJUSTMENT) & Used(SCRAP, USED AND SECONDHAND GOODS):
"†† The scrap; used and secondhand goods; noncomparable imports; 
and rest of the world adjustment lines identify special commodities which may appear
in the commodity dimension of the make and use tables but which do not appear 
in the industry dimension."(c) Use_SUT_Framework_2007_2012_DET's footnote'''

# Read aggregated 2007 73*71
aggr2007 = pd.read_excel('use2007aggr.xls',index_col=0, skiprows=5)

# Read codes
codes = pd.read_excel('Use_SUT_Framework_2007_2012_DET.xlsx', sheet_name='NAICS Codes', skiprows=4, dtype=str)

# Get codes of 73 industries
mid_aggr_codes = aggr2007.index[1:-9]
mid_aggr_codes
```

```python
mid_min_dict = {}
# Get indexes of 71 industries codes
mid_code_index = codes.index[codes.Summary.isin(mid_aggr_codes)]
mid_code_index = np.append(mid_code_index, codes.shape[0])

# Get 405 industries codes
for i, mid_code in enumerate(mid_aggr_codes):
    group = codes.loc[mid_code_index[i]:mid_code_index[i+1]-1,'Detail']
    min_codes = [] 
    # Filter not nan codes from table
    for code in group:
        if code != 'nan' and len(code) <= 6:
            min_codes.append(code)
    mid_min_dict[mid_code] = min_codes  
mid_min_dict
```
